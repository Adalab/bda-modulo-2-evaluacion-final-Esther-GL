```python
import requests
import pandas as pd
```

```python
url_pelis = f"https://beta.adalab.es/resources/apis/pelis/pelis.json"

llamada = requests.get(url_pelis)

print(llamada.status_code)
```

```python
llamada.json()
```

```python
datos_peli = llamada.json()
```

```python
datos_peli[0].keys()
```

```python
datos_ejercicio_peliculas = []

for pelicula in datos_peli:
    datos_ejercicio_peliculas.append({
    "titulo": pelicula.get("titulo", "unknown"),
    "año": pelicula.get("año", "unknown"),
    "duracion": pelicula.get("duracion", "unknown"),
    "genero": pelicula.get("genero", "unknown"),
    "adultos": pelicula.get("adultos", "unknown"),})

```

```python
dataframe_peli = pd.DataFrame(datos_ejercicio_peliculas)

dataframe_peli
```
